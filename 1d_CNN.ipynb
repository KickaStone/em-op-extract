```python
import matplotlib.pyplot as plt
from test import test_model
from models import CNN_1d
import os

model_cnn = CNN_1d(input_size=131072, num_classes=9)
labels = ["aes", "sha1prf", "hmacsha1", "des_openssl", "aes_openssl", "aes_tiny", "sha1", "sha1transform", "noise"]
models_root = "./models"
model_name = "best_cnn-epoch-{}.pt"
datasets = ['nodemcu-random-test2']
```

```python
# get classification results
classification_reports = []
for i in range(10):
    model_path = os.path.join(models_root, model_name.format(i))
    model_report = test_model(datasets, model_path, model_cnn)
    classification_reports.append(model_report)
print('Test Done', 'Get {} reports'.format(len(classification_reports)))
```

```python
# plot test result
acc = []
avg_acc = []
for i in range(len(labels)):
    acc.append([])
    
for i, rp in enumerate(classification_reports):
    for k in range(len(labels)):
        acc[k].append(rp[f'{k}']['precision'])
    avg_acc.append(rp['accuracy'])

plt.figure(figsize=(10, 8))
for i, lb in enumerate(labels):
    plt.plot(acc[i], label=lb)
plt.plot(avg_acc, label="avg_acc", )
plt.legend()
plt.title('CNN1d Classification Report')
plt.show()
```
